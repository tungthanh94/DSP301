```python
import pandas as pd
import numpy as np
```

```python
file = None
while file is None:
    try:
        
        n = input('Enter a class to grade (i.e. class1 for class1.txt): ')
        
        
        file = open(f'Data Files/{n}.txt', 'r')
        
        print(f'Successfully opened {n}')
        print('**** ANALYZING ****')
    except:
        print('File cannot be found.\nPlease try again!')
valid = 0
invalid = 0
valid_line =[]
for i in file:
    line = i.strip().split(',')
    
    if len(line) != 26:
        print('Invalid line of data: does not contain exactly 26 values:')
        print(i)
        invalid += 1
        
    elif len(line[0]) != 9 or not line[0].startswith('N') or not line[0][1:].isnumeric():
        print('Invalid line of data: N# is invalid')
        print(i)
        invalid += 1
    else:
        valid += 1
        valid_line.append(line)
print('**** REPORT ****')
print('Total valid lines of data:', valid)
print('Total invalid lines of data:', invalid)

answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"
```
